```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = {
    'FirstName' : ['Shreyas','Aditya','Sakshi','Rutuja','Yuga','Vasu','Omkar','Sanket','Gauri','Subodh','Aryan'],
    'LastName' : ['Joshi','Kulkarni','Wagh','Deshpande','Kulkarni','Kulkarni','Katkar','Mandekar','Sathe','Joshi','Gole'],
    'Gender' : ['Male','Male','Female','Female','Female','Female','Male','Male','Female','Male','Male'],
    'WT' : [23,'na',12,34,28,14,19,23,'-','nan',31],
    'AI' : [19,15,21,25,29,19,26,'-',21,10,'na'],
    'CC' : [23,19,29,56,'-',11,10,'na',19,15,21],
    'DSBDA' : [13,14,19,23,'-','nan','na',12,34,'nan',31]
}
```

```python
df = pd.DataFrame(dataset)
```

```python
df
```

```python
missing_value_format = ['na','nan','-']
```

```python
df.replace(missing_value_format,np.nan,inplace = True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
print(df.dropna(axis=0))
```

```python
print(df.dropna(axis=1))
```

```python
df["WT"].fillna(0,inplace = True)
df
```

```python
df['AI'].fillna(method = 'pad',inplace = True)
df
```

```python
df['CC'].fillna(method = 'bfill',inplace = True)
df
```

```python
mean_ds = df['DSBDA'].mean()
df['DSBDA'].fillna(mean_ds,inplace = True)
df
```

```python
df["WT"] = df["WT"].astype(int)
df["DSBDA"] = df["DSBDA"].astype(int)
df["CC"] = df["CC"].astype(int)
df["AI"] = df["AI"].astype(int)
```

```python
df
```

```python
sns.boxplot(x=df['WT'])
plt.show()
```

```python
sns.boxplot(x=df['DSBDA'])
plt.show()
```

```python
sns.boxplot(x=df['CC'])
plt.show()
```
