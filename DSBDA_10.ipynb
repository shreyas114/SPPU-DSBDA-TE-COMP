```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('iris')
```

```python
sns.boxplot(data['sepal_length'])
```

```python
sns.boxplot(data['petal_width'])
```

```python
fig,axes = plt.subplots(2,2,figsize=(16,10))
sns.histplot(data["sepal_length"],ax=axes[0,0],kde=True)
sns.histplot(data["sepal_width"],ax=axes[0,1],kde=True)
sns.histplot(data["petal_length"],ax=axes[1,0],kde=True)
sns.histplot(data["petal_width"],ax=axes[1,1],kde=True)
```

```python
fig,axes = plt.subplots(2,2,figsize=(16,10))
sns.boxplot(y=data["sepal_length"],x=data["species"],data=data,ax=axes[0,1])
sns.boxplot(y=data["sepal_width"],x=data["species"],data=data,ax=axes[1,0])
sns.boxplot(y=data["petal_length"],x=data["species"],data=data,ax=axes[1,1])
sns.boxplot(y=data["petal_width"],x=data["species"],data=data,ax=axes[0,0])
```
